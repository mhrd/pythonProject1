```python
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM
from sklearn.preprocessing import MinMaxScaler

# Simulating some stock price data
prices = np.array([100, 101, 101, 102, 100, 99, 100, 101, 102, 101])

# Reshape and scale the data
prices = prices.reshape(-1, 1)
scaler = MinMaxScaler(feature_range=(0, 1))
prices_scaled = scaler.fit_transform(prices)

# Create sequences of 3 prices to predict the next price
X = []
y = []
n_future = 1  # number of days we want to predict into the future
n_past = 3     # number of past days we want to use to predict the future

for i in range(n_past, len(prices_scaled) - n_future +1):
    X.append(prices_scaled[i - n_past:i, 0:prices_scaled.shape[1]])
    y.append(prices_scaled[i + n_future - 1:i + n_future, 0])

X, y = np.array(X), np.array(y)

# Creating a model
model = Sequential()
model.add(LSTM(units=50, return_sequences=True, input_shape=(X.shape[1], X.shape[2])))
model.add(LSTM(units=50))
model.add(Dense(units=1))

model.compile(optimizer='adam', loss='mean_squared_error')

# Training the model
model.fit(X, y, epochs=1, batch_size=1)

# Making a prediction for the 11th value
x_input = prices_scaled[-n_past:]  # last 3 values
x_input = x_input.reshape((1, n_past, 1))
predicted_price = model.predict(x_input)
predicted_price = scaler.inverse_transform(predicted_price)

print("Predicted price for the 11th day:", predicted_price[0, 0])

# Adding the predicted price to the plot
prices_with_prediction = np.append(prices, predicted_price)
plt.plot(prices, label='Original Price')
plt.plot(prices_with_prediction, label='Original Price + Prediction', linestyle='--')
plt.legend()
plt.show()
```
